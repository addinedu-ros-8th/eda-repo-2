```python
import pandas as pd

usec = [2, 10, 19, 49, 63, 66, 71, 72, 73, 74, 76, 98, 102]

df_ads = pd.read_csv("../data/ads.csv", usecols = usec)
df_adas = pd.read_csv("../data/adas_level2.csv", usecols = usec)
```

```python
# Weather 기본값 설정
df_ads["Weather"] = "Other"

# 조건에 따라 Weather 열 수정
df_ads.loc[df_ads["Weather - Clear"] == 'Y', "Weather"] = "Clear"
df_ads.loc[df_ads["Weather - Snow"] == 'Y', "Weather"] = "Snow"
df_ads.loc[df_ads["Weather - Cloudy"] == 'Y', "Weather"] = "Cloudy"
df_ads.loc[df_ads["Weather - Rain"] == 'Y', "Weather"] = "Rain"

# 불필요한 열 삭제
df_ads = df_ads.drop(columns=["Weather - Clear", "Weather - Snow", "Weather - Cloudy", "Weather - Rain"])

# NaN 행 삭제
df_ads = df_ads.dropna()
```

```python
# Weather 기본값 설정
df_adas["Weather"] = "Other"

# 조건에 따라 Weather 열 수정
df_adas.loc[df_adas["Weather - Clear"] == 'Y', "Weather"] = "Clear"
df_adas.loc[df_adas["Weather - Snow"] == 'Y', "Weather"] = "Snow"
df_adas.loc[df_adas["Weather - Cloudy"] == 'Y', "Weather"] = "Cloudy"
df_adas.loc[df_adas["Weather - Rain"] == 'Y', "Weather"] = "Rain"

df_adas.loc[df_adas["ADAS/ADS System Version"] == "[REDACTED, MAY CONTAIN CONFIDENTIAL BUSINESS INFORMATION]", "ADAS/ADS System Version"] = "level 2"

# 불필요한 열 삭제
df_adas = df_adas.drop(columns=["Weather - Clear", "Weather - Snow", "Weather - Cloudy", "Weather - Rain"])

# NaN 행 삭제
df_adas = df_adas.dropna()
```

```python
# Tesla, Inc. 행 필터링
tesla_df = df_adas[df_adas["Reporting Entity"] == "Tesla, Inc."]

# Tesla 전체 수
total_tesla = len(tesla_df)

# Tesla 중 Weather가 "Other"인 수
tesla_other_weather = len(tesla_df[tesla_df["Weather"] == "Other"])

print(f"Total Tesla rows: {total_tesla}")
print(f"Tesla rows with 'Weather' as 'Other': {tesla_other_weather}")

tesla_unknown = len(tesla_df[tesla_df["SV Pre-Crash Movement"] == "Unknown"])
print(tesla_unknown)
```
