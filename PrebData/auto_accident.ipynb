```python
import pandas as pd

usec = [2, 49, 61, 66, 67, 72, 73, 74, 76, 98, 102]

df_ads = pd.read_csv("../data/ads.csv", usecols = usec)
df_adas = pd.read_csv("../data/adas_level2.csv", usecols = usec)
print(df_ads)
```

```python
# Weather 기본값 설정
df_ads["Weather"] = "Other"

# 조건에 따라 Weather 열 수정
df_ads.loc[df_ads["Weather - Clear"] == 'Y', "Weather"] = "Clear"
df_ads.loc[df_ads["Weather - Snow"] == 'Y', "Weather"] = "Snow"
df_ads.loc[df_ads["Weather - Cloudy"] == 'Y', "Weather"] = "Cloudy"
df_ads.loc[df_ads["Weather - Rain"] == 'Y', "Weather"] = "Rain"

# 불필요한 열 삭제
df_ads = df_ads.drop(columns=["Weather - Clear", "Weather - Snow", "Weather - Cloudy", "Weather - Rain"])

# NaN 행 삭제
df_ads = df_ads.dropna()



# Weather 기본값 설정
df_adas["Weather"] = "Other"

# 조건에 따라 Weather 열 수정
df_adas.loc[df_adas["Weather - Clear"] == 'Y', "Weather"] = "Clear"
df_adas.loc[df_adas["Weather - Snow"] == 'Y', "Weather"] = "Snow"
df_adas.loc[df_adas["Weather - Cloudy"] == 'Y', "Weather"] = "Cloudy"
df_adas.loc[df_adas["Weather - Rain"] == 'Y', "Weather"] = "Rain"

# 불필요한 열 삭제
df_adas = df_adas.drop(columns=["Weather - Clear", "Weather - Snow", "Weather - Cloudy", "Weather - Rain"])
#df_adas.loc[df_adas["ADAS/ADS System Version"] == "[REDACTED, MAY CONTAIN CONFIDENTIAL BUSINESS INFORMATION]", "ADAS/ADS System Version"] = "level 2"

# NaN 행 삭제
df_adas = df_adas.dropna()
```

```python
df_ads["Speed"] = df_ads["SV Precrash Speed (MPH)"] * 1.60934
df_ads["Speed"] = df_ads["Speed"].round().astype(int)
df_ads = df_ads.drop(columns=["SV Precrash Speed (MPH)"])

df_ads.rename(columns={"Reporting Entity" : "Brand"}, inplace=True)
df_ads.rename(columns={"ADAS/ADS System Version" : "Level"}, inplace=True)
df_ads.rename(columns={"Incident Date": "Accident date"}, inplace=True)
df_ads.rename(columns={"Roadway Type": "Road"}, inplace=True)
df_ads.rename(columns={"Roadway Surface": "Pavement"}, inplace=True)
df_ads.rename(columns={"SV Pre-Crash Movement": "Movement"}, inplace=True)

df_ads.head(5)
```

```python
df_adas["Speed"] = df_adas["SV Precrash Speed (MPH)"] * 1.60934
df_adas["Speed"] = df_adas["Speed"].round().astype(int)
df_adas = df_adas.drop(columns=["SV Precrash Speed (MPH)"])

df_adas.rename(columns={"Reporting Entity" : "Brand"}, inplace=True)
df_adas.rename(columns={"ADAS/ADS System Version" : "Level"}, inplace=True)
df_adas.rename(columns={"Incident Date": "Accident date"}, inplace=True)
df_adas.rename(columns={"Roadway Type": "Road"}, inplace=True)
df_adas.rename(columns={"Roadway Surface": "Pavement"}, inplace=True)
df_adas.rename(columns={"SV Pre-Crash Movement": "Movement"}, inplace=True)

df_adas.head(5)
```
