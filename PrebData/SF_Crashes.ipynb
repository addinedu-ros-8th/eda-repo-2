```python
import pandas as pd
```

## Get the data (CSV) file

```python
dataSF = pd.read_csv("../data/CrahsesSF.csv", 
                   usecols=["CASE_ID", "COLLISION_DATE", "COUNTY", "INTERSECTION", "WEATHER_1", "STATE_HWY_IND", "ROAD_SURFACE", "CONTROL_DEVICE", "PCF_VIOL_CATEGORY"])

dataSF
```

## Replace ABCDE to our index (012345)

```python
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("A", 0)
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("B", 1)
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("C", 2)
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("D", 3)
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("E", 4)
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("F", 4)
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("-", 4)
dataSF["WEATHER_1"] = dataSF["WEATHER_1"].replace("G", 4)

dataSF["ROAD_SURFACE"] = dataSF["ROAD_SURFACE"].replace("A", 0)
dataSF["ROAD_SURFACE"] = dataSF["ROAD_SURFACE"].replace("B", 1)
dataSF["ROAD_SURFACE"] = dataSF["ROAD_SURFACE"].replace("C", 2)
dataSF["ROAD_SURFACE"] = dataSF["ROAD_SURFACE"].replace("D", 3)
dataSF["ROAD_SURFACE"] = dataSF["ROAD_SURFACE"].replace("-", 3)
dataSF["ROAD_SURFACE"] = dataSF["ROAD_SURFACE"].replace("H", 3)

dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("01", 6)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("02", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("03", 2)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("04", 2)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("05", 0)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("06", 1)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("07", 1)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("08", 3)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("09", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("10", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("11", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("12", 9)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("13", 5)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("14", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("15", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("16", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("17", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("18", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("19", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("20", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("21", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("22", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("23", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("24", 7)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("00", 8)
dataSF["PCF_VIOL_CATEGORY"] = dataSF["PCF_VIOL_CATEGORY"].replace("- ", 8)

dataSF
```

## Remake with proper columns & type to save into Data Base

```python
data = pd.DataFrame(columns=["accident_id", "accident_date", "speed",  "age", "sex", "auto_id",  "state_id", "weather_id",
                             "road_id", "movement_id", "brand_id","pavement_id"])

data.accident_id = dataSF["CASE_ID"]      
data.state_id = 1
data.accident_date = dataSF["COLLISION_DATE"]
data.auto_id = 1
data.brand_id = None
data.weather_id = dataSF["WEATHER_1"]
data.pavement_id = dataSF["ROAD_SURFACE"]
data.movement_id = dataSF["PCF_VIOL_CATEGORY"]
data.speed = None
data.age = None
data.sex = None
data.road_id =  dataSF.apply(lambda row: '2' if row['INTERSECTION'] == 'Y' 
                         else ('1' if row['STATE_HWY_IND'] == 'Y' 
                               else '0'), axis=1)
data
```

```python
data = data.drop(columns=["accident_id"])
```
