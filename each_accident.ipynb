```python
efficient_data = 기존_데이터.loc[:,['발생년월','가해운전자 연령대','가해운전자 성별','법규위반','노면상태','기상상태','도로형태']]
efficient_data.head()
```

```python
# 매핑 기준 정의
pavement_mapping = {
    '건조': 0,
    '젖음/습기': 1,
    '서리/결빙': 2,
    '적설': 2,
    '해빙': 2,
    '기타': 3,
    '침수': 3
}

# '노면상태' 컬럼의 값을 매핑
efficient_data['노면상태'] = efficient_data['노면상태'].map(pavement_mapping)

# 결과 확인
efficient_data.tail(10)
```

```python
# 매핑 기준 정의
weather_mapping = {
    '맑음': 0,
    '흐림': 1,
    '안개': 1,
    '비': 2,
    '눈': 3,
    '기타': 4
}

# '기상상태' 컬럼의 값을 매핑
efficient_data['기상상태'] = efficient_data['기상상태'].map(weather_mapping)

# 결과 확인
efficient_data.tail(10)
```

```python
# 매핑 기준 정의
road_type_mapping = {
    '단일로 - 교량': 0,
    '단일로 - 지하차도(도로)내': 0,
    '단일로 - 고가도로위': 0,
    '단일로 - 터널': 0,
    '단일로 - 기타': 1,
    '교차로 - 교차로안': 2,
    '교차로 - 교차로부근': 2,
    '교차로 - 교차로횡단보도내': 2,
    '주차장 - 주차장': 3,
    '기타 - 기타': 4,
    '미분류 - 미분류': 4
}

# '도로형태' 컬럼의 값을 매핑
efficient_data['도로형태'] = efficient_data['도로형태'].map(road_type_mapping)

# 결과 확인
efficient_data.tail(10)
```

```python
# NaN 값 확인
print(efficient_data[efficient_data['도로형태'].isna()])
```

```python
# 매핑 기준 정의
violation_mapping = {
    '중앙선침범': 0,
    '차로위반': 1,
    '교차로운행방법위반': 1,
    '과속': 2,
    '안전거리미확보': 2,
    '직진우회전진행방해': 3,
    '불법유턴': 4,
    '음주': 6,
    '기타': 8,
    '보행자보호의무위반': 8,
    '안전운전불이행': 8,
    '신호위반': 9
}

# '법규위반' 컬럼 매핑
efficient_data['법규위반'] = efficient_data['법규위반'].map(violation_mapping)

# 결과 확인
efficient_data.tail(10)
```

```python
efficient_data["가해운전자 연령대"].value_counts()
```

```python
# 매핑 기준 정의
age_mapping = {
    '65세 이상': 65,
    '61-64세':62,
    '51-60세': 55,
    '41-50세': 45,
    '31-40세': 35,
    '21-30세': 25,
    '20세 이하': 20,
    '기타불명': None
}

# '법규위반' 컬럼 매핑
efficient_data['가해운전자 연령대'] = efficient_data['가해운전자 연령대'].map(age_mapping)

# 결과 확인
efficient_data.tail(10)
```

```python
# 매핑 기준 정의
sex_mapping = {
    '남': 'M',
    '여': 'F'
}

# '법규위반' 컬럼 매핑
efficient_data['가해운전자 성별'] = efficient_data['가해운전자 성별'].map(sex_mapping)

# 결과 확인
efficient_data.tail(10)
```

```python
# 새로운 컬럼 추가 및 값 할당
efficient_data['state'] = 0  # 'state' 컬럼에 0 추가
efficient_data['speed'] = None  # 'speed' 컬럼에 null 값 추가
efficient_data['brand'] = None  # 'brand' 컬럼에 null 값 추가
efficient_data['auto'] = 1  # 'auto' 컬럼에 1 추가

# 결과 확인
efficient_data.head()
```
